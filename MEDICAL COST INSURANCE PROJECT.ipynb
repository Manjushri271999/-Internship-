```python
plt.figure(figsize=(5,5))
sns.countplot(x='sex', data=icost)
plt.title('Gender Distribution')
plt.show( )
```

```python
sns.countplot(x='smoker', data=icost)
plt.title('smoker')
plt.show( )
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='region',data=icost)
plt.title('Region')
plt.show( )
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='region',y='charges',data=icost)
plt.title('Cost vs Region ')
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='smoker',y='charges',data=icost)
plt.title('Charges for Smokers')
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='sex', y='charges',hue='smoker',data=icost)
plt.title('Charges for Smokers')
```

```python
fig, axes= plt.subplots(1,3, figsize=(15,5), sharey= True)
fig.suptitle('Visualizing categorial columns ')
sns.boxenplot(x='smoker', y='charges',data=icost,ax=axes[0])
sns.boxenplot(x='sex', y='charges',data=icost,ax=axes[1])
sns.boxenplot(x='region', y='charges',data=icost,ax=axes[2])
```

```python
icost[['age','bmi','children','charges']].hist(bins=30, figsize=(10,10), color='blue')
plt.show( )
```

```python
icost['sex'] = icost['sex'].apply({'male':0, 'female':1}.get)
icost['smoker'] = icost['smoker'].apply({'yes':1, 'no':0}.get)
icost['region'] = icost['region'].apply({'southwest':1, 'southeast':2, 'northwest':3, 'northeast':4}.get)
```
